```python
# Make a simple plot
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Generate a fake model and data
xmodel = np.arange(20)
ymodel = 3*xmodel+2
xdata = xmodel + 0.9*(np.random.rand(20)-0.5)
ydata = ymodel + 0.5*(np.random.rand(20)-0.5)

# Make a very simple plot of the data and the model
plt.plot(xmodel,ymodel)
plt.scatter(xdata,ydata)
```

**Let's get pretty!**

Now make a more customized plot. The third argument in a plot command, e.g. 'k-' in the first example below, specifies the color and the style of the symbols: the first letter specifies the color ("r"=red, "b"=blue, "g"=green, "k"=black, etc.) and the second symbol specifies the symbol/line type ("-"=solid line, ":"=dotted line, "o"=circle, "s"=square, etc.). The "label" keyword is used when making the legend. You can add in "lw=X" or "ms=Y" to adjust
 the linewidth and symbol size. 
 
 Type "help(plot)" for further help.

```python
plt.plot(xmodel,ymodel,'k-',label='this is the model')
plt.plot(xdata,ydata,'ro',label='this is the data')

# The following commands adjust the ranges of the axes, label the
# axes, create a legend, and create a title. You can also add in
# "fontsize=X" to any of the commands that generate text.
plt.axis([0,20,0,60])
plt.xlabel('Time [days]')
plt.ylabel('Flux [erg/s/cm^2/s')
plt.legend(loc='upper left')
plt.title('A simple plot')

# you can also save your figures to file
plt.savefig('exampleplot.png',dpi=300, bbox_inches='tight')
```
