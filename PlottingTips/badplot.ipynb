```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# First make a bad plot
plt.clf()
data1 = np.loadtxt('data/data2.txt')
xdata1 = data1[:,0]
ydata1 = data1[:,1]
plt.plot(xdata1, ydata1,'ko',markersize=1)

data2 = np.loadtxt('data/data3.txt')
xdata2 = data2[:,0]
ydata2 = data2[:,1]
plt.plot(xdata2, ydata2,'ko',markersize=1)

model = np.loadtxt('data/model.txt')
xmodel = model[:,0]
ymodel = model[:,1]
plt.plot(xmodel, ymodel,'k-',linewidth=0.5)

plt.axis([-50,50,-100,100])
plt.ylabel('flux',fontsize=9)
```

```python
# These commands will make it look more reasonable
plt.axis([0,9,10,70])
plt.xlabel('time [days]',fontsize=12)
plt.ylabel('flux [ergs/s]',fontsize=12)
plt.plot(xdata1,ydata1,'bo',markersize=6,label='telescope 1')
plt.plot(xdata2,ydata2,'ro',markersize=6,label='telescope 2')
plt.plot(xmodel,ymodel,'k-',linewidth=1,label='model')
plt.legend(loc='upper left')
```
