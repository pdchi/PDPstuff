# Gaussian fitting example

The Gaussian is a common distribution often found in nature (like spectral lines...), given by...

\begin{equation*}
\mathcal{G} (x) = I * \text{exp}\left({-\frac{(x-\mu)^2 }{2\sigma^2}}\right)
\end{equation*}

where $~I~$ is the height of the gaussian, $~\mu~$ is the mean or center of the distribution, and $~\sigma~$ is the variance of the distribution. 

---

For a double gaussian of the same mean and variance, we can simply define them as 

\begin{equation*}
\mathcal{D} (x) = \mathcal{G}_1(x) + \mathcal{G}_2(x) = I * \text{exp}\left({-\frac{(x-\mu)^2 }{2\sigma^2}}\right) + I * \text{exp}\left({-\frac{(x-\mu - p_{sep})^2 }{2\sigma^2}}\right)
\end{equation*}

with a seperation $~p_{sep}$. 

---

There are also some additional quantities we can define:

The Full Width at Half Max is the span of the two points on either side of the peak that are half of the peak's value, with respect to some continuum. For a gaussian distribution, the FWHM is related directly to the variance by $~FWHM = 2\sqrt{2\text{ln}2}\sigma\approx2.355\sigma$

![title](500px-FWHM.png)

The Equivalent Width is the width of a rectangle with the same area under the curve of the gaussian, with respect to some continuum.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# lets import our own gaussian fitting routine
from gaussian_fitter import gaussfit
```

```python
# lets try it out!
# generate some fake data
x = np.linspace(-3,3,1000)
mean = 1.2
sig = .7
y = 8. * np.exp(-(x - mean)**2./(2. * sig**2.0))
y = y + np.random.normal(size=len(x))
plt.plot(x,y)
```

```python
gaussfit(x,y)
```

```python
# you can store the output of guassfit into a variable directly
# the guassfit function also has some extra keywords
fit = gaussfit(x,y, quiet=True, graph=True)
print(fit)
```
